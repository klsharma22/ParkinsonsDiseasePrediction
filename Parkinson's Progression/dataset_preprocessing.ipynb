```python
import pandas as pd
```

```python
peptides = pd.read_csv("train_peptides_1.csv")
proteins = pd.read_csv("train_proteins_1.csv")
```

```python
peptides.drop(columns= ["Unnamed: 0"], inplace= True)
```

```python
peptides
```

```python
proteins.drop(columns= ['Unnamed: 0'], inplace= True)
```

```python
peptides.columns[1:]
```

```python
def replace_nan(df: pd.DataFrame, columns: list):
    for column in columns:
        df[column].fillna(df[column].mean(), inplace= True)
    return df
```

```python
peptides_cleaned = replace_nan(peptides)
```

```python
proteins_cleande = replace_nan(proteins)
```

```python
for column in proteins_cleande.columns[1:]:
    for value in proteins_cleande[column].isna():
        if value:
            print(f"Nan is present in {column}")
```

```python
proteins_cleande.isna().value_counts()
```

```python
peptides_cleaned.to_csv("train_peptides_1.csv")
proteins_cleande.to_csv("train_proteins_1.csv")
```

```python
new_peptides = pd.read_csv("train_peptides_1.csv")
new_peptides.isna().value_counts()
```

```python
def create_visit_month(df: pd.DataFrame):
    months = []
    for visit_id in df['visit_id']:
        visit_month = visit_id.split('_')[-1]
        months.append(visit_month)

    df.insert(2, column= 'visit_month', value= months)
```

```python
create_visit_month(new_peptides)
```

```python
new_peptides.to_csv("train_peptides_1.csv")
```

```python
new_proteins = pd.read_csv('train_proteins_1.csv')
```

```python
create_visit_month(new_proteins)
```

```python
new_proteins.to_csv('train_proteins_1.csv')
```
